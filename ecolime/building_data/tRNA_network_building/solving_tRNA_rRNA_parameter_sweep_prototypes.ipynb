```python
import os
def open_and_solve_prototypes(path, using_soplex = False):
    master_dict = {}
    from tqdm import tqdm
    
    for filename in tqdm(os.listdir(path)):
        #print filename
        with open(path + filename, 'rb') as infile:
            me = cPickle.load(infile)
            output_dict = {}
            
        

            if using_soplex:
                binary_search(me, min_mu=.1, max_mu=.3, debug=False, mu_accuracy=1e-2)
            else:
                from qminos.me1 import ME_NLP1
                me.reactions.EX_glc__D_e.lower_bound = -1000.
                me.reactions.EX_glc__D_e.upper_bound = 1000.
                # The object containing solveME methods--composite that uses a ME model object 
                # Provide growth_key = 'mu' for minime models,
                me_nlp = ME_NLP1(me, growth_key='mu')
                # Use bisection for now (until the NLP formulation is worked out for the new prototype 44
                muopt, hs, xopt, cache = me_nlp.bisectmu(precision=1e-6, mumax=1.)    
                # Access the solution that is saved in the original cobrame object
                sol = me.solution
                sol.f
                sol.x_dict
                
           
            output_dict.update({'sol' : sol})
            output_dict.update({'sol_f' : sol.f})
            output_dict.update({'sol_xdict' : sol.x_dict})
            output_dict.update({'metabolic_flux' : me.get_metabolic_flux()})
            output_dict.update({'transcription_flux' : me.get_transcription_flux()})
            output_dict.update({'translation_flux' : me.get_translation_flux()})
            
        master_dict.update({filename : output_dict})
    return master_dict
```

```python
import pandas as pd
control_prototype = './RNA_sweep_control_prototypes/proto_65_NO_tRNA_mods_NO_rRNA_mods.pickle'


control_df = pd.DataFrame()
control_df.loc[ 0 , 'rRNA_mods'] = True
control_df.loc[ 1 , 'rRNA_mods'] = True
control_df.loc[ 2 , 'rRNA_mods'] = False
control_df.loc[ 3 , 'rRNA_mods'] = False

control_df.loc[ 0 , 'tRNA_mods'] = True
control_df.loc[ 1 , 'tRNA_mods'] = False
control_df.loc[ 2 , 'tRNA_mods'] = True
control_df.loc[ 3 , 'tRNA_mods'] = False

control_df
```

```python
def open_and_solve_control_prototypes(control_prototype, control_df,  using_soplex = False):
    import copy
    from copy import deepcopy
    from tqdm import tqdm
    master_dict = {}
    
    for index, row in tqdm(control_df.iterrows()):
        print index
        with open(control_prototype, 'rb') as infile:
            me = cPickle.load(infile)

            #ribosome no rRNA modifications
            if row.rRNA_mods == False:
                rRNA_name = 'NO'
                # Changed verbose=False to squash complex creation output
                ribosome.add_ribosome(me, verbose=False, rRNA_mods = False)

            #ribosome with all rRNA modifications
            elif row.rRNA_mods == True:
                rRNA_name = 'YES'
                # Changed verbose=False to squash complex creation output
                ribosome.add_ribosome(me, verbose=False, rRNA_mods = True)

            # These should be coupled to translation not ribosome formation per iOL1650
            Inf_modifications = ['Translation_initiation_factor_InfA', 'Translation_gtp_initiation_factor_InfB',
                        'Translation_initiation_factor_InfC']
            for sub in Inf_modifications:
                me.complex_data.ribosome.modifications[sub] = 0
                for data in me.translation_data:
                    data.modifications[sub]= -1

            #tRNAs fully modified
            if row.tRNA_mods == True:
                tRNA_name = 'YES'
                pass

            # no tRNA modifications
            elif row.tRNA_mods == False:
                tRNA_name = 'NO'
                for trna_charging_rxn in me.reactions.query('charging_tRNA'):
                    trna_charging_rxn.tRNAData.modifications = {}


            me.update()
            me.prune()

            output_dict = {}

            if using_soplex:
                binary_search(me, min_mu=.1, max_mu=.3, debug=False, mu_accuracy=1e-2)
            else:
                from qminos.me1 import ME_NLP1
                me.reactions.EX_glc__D_e.lower_bound = -1000.
                me.reactions.EX_glc__D_e.upper_bound = 1000.
                # The object containing solveME methods--composite that uses a ME model object 
                # Provide growth_key = 'mu' for minime models,
                me_nlp = ME_NLP1(me, growth_key='mu')
                # Use bisection for now (until the NLP formulation is worked out for the new prototype 44
                muopt, hs, xopt, cache = me_nlp.bisectmu(precision=1e-6, mumax=1.)    
                # Access the solution that is saved in the original cobrame object
                sol = me.solution
                sol.f
                sol.x_dict


            output_dict.update({'sol' : sol})
            output_dict.update({'sol_f' : sol.f})
            output_dict.update({'sol_xdict' : sol.x_dict})
            output_dict.update({'metabolic_flux' : me.get_metabolic_flux()})
            output_dict.update({'transcription_flux' : me.get_transcription_flux()})
            output_dict.update({'translation_flux' : me.get_translation_flux()})

        proto_name = 'proto_65_tRNAmods_' +  tRNA_name + '_rRNAmods_' + rRNA_name + '.pickle'
        master_dict.update({proto_name : output_dict})

    return master_dict

        
```

```python
import copy
from copy import deepcopy
control_results = copy.deepcopy(open_and_solve_control_prototypes(control_prototype, control_df,  using_soplex = False))
control_results
```
